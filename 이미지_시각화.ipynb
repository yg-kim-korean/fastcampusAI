#이미지 시각화

```
#패키지 불러오기
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

%matplotlib inline
```

```
#이미지 파일열기
path='/content/drive/MyDrive/Colab Notebooks/1.jpg'
image_pil = Image.open(path)
image = np.array(image_pil)
```

```
image.shape
```

```
#이미지 들여다보기
np.min(image), np.max(image)
```

```
#그래프로 시각화 하기
plt.hist(image.ravel(),256,[0,256])
```

```
plt.imshow(image)
plt.show()
```

```
#흑백으로 열기
path
```

```
image_pil = Image.open(path).convert("L")
image_bw = np.array(image_pil)
```

```
image_bw.shape
```

```
#흑백 이미지 열기
plt.imshow(image_bw, 'gray')
plt.show()
```

```
#다른색상으로 cmap 표현하기
#gray scale
image_bw.shape
```

```
plt.imshow(image_bw,'gray')
plt.show()
```

```
#RdBu
plt.imshow(image_bw,'RdBu')
plt.show()
```

```
#jet
plt.imshow(image_bw, 'jet')
plt.show()
```

#Colorbar  추가하기


```
plt.imshow(image_bw,'jet')
plt.colorbar()
plt.show()
```

#이미지 설정


```
#이미지 보기 사이즈 조절
plt.figure(figsize=(10,10))
plt.imshow(image)
plt.show()
```
